```python
# Python Libreries
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
from sql_queries import read_sql
```

```python
# Read Data from the States Table
teams_matches_df = read_sql("""--sql
    SELECT * FROM teams_matches
""")
```

```python
teams_matches_df.describe()
```

```python
teams_matches_df.columns
```

```python
targets = ['duration', 'points', 'rebounds', 'assists', 'steals', 'blocks',
       'turnovers']
for col in targets:
    fig = go.Figure(data = go.Histogram(x=teams_matches_df[col]))
    fig.layout.update(title=col)
    fig.show()
```

```python
go.Figure(data=go.Box(y=teams_matches_df.duration), layout=dict(title='duration'))
```
